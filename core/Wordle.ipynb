```python
import re
import pandas as pd
```

```python
df = pd.read_csv("unigram_freq.csv", header=None, delimiter=",")
```

```python
df
```

```python
df2 = [df[0][x] for x in range(len(df[0])) if len(str(df[0][x])) == 5]
```

```python
search_regex = r"[a-z][a-z][t][a-z][e]"
```

```python
possibilities = []
```

```python
for x in df2:
    check = re.fullmatch(search_regex, x)
    if check:
        possibilities.append(check.group())
```
